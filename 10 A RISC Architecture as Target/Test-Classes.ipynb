```python
compileString(src = """class Person
    var sin, age: integer
    var isMale: boolean
    procedure setSin(self, sin : integer)
        self.sin := sin
    procedure setAge(self, age : integer)
        self.age := age
    procedure printSIN(self)
        write(self.sin); writeln()
    procedure printAge(self)
        write(self.age); writeln()
class Student extends Person
    var grade : integer
    procedure setGrade(self, grade : integer)
        self.grade := grade
    procedure printGrade(self)
        write(self.grade); writeln()
class TBStudent extends Student
    var midterm1 : integer
    var passed : boolean
    procedure setMidterm1(self, mid : integer)
        self.midterm1 := mid
    procedure setPass(self)
        if self.midterm1 > 60 then self.passed := true else self.passed := false
object j : TBStudent
procedure notMale()
    j.isMale := false
procedure minusGrade(grade : integer)
    j.grade := j.grade - grade
program myProg
    j.setSin(919)
    j.setAge(20)
    j.printSIN()
    j.printAge()
    j.setGrade(100)
    j.printGrade()
    j.setMidterm1(59)
    j.setPass()
    if j.passed then write(888) else write(555)
""", dstfn = "AssemblyFiles/RISCV-class2.s",target = 'riscv')
# 919
# 20
# 100
# 555
```

```python
compileString(src="""class Veichle
    var price, color, gas : integer
    procedure init(self, price : integer, color : integer)
        self.color := color
        self.price := price
        self.gas := 0
    procedure fillUpTank(self)
        self.gas := 100
    procedure emptyTank(self)
        self.gas := 0
    procedure gasLeft(self)
        write(self.gas); writeln()
    procedure showStats(self)
        write(self.color); writeln()
        write(self.price); writeln()
        write(self.gas); writeln()
    
class Truck extends Veichle
    var tires : integer
    procedure init(self, price : integer, color : integer, tires : integer)
        self.color := color
        self.price := price
        self.gas := 0
        self.tires := tires
    procedure beep(self)
        write(8888); writeln()
    procedure showStats(self)
        write(self.color); writeln()
        write(self.price); writeln()
        write(self.gas); writeln()
        write(self.tires); writeln()

class Car extends Veichle
    var speed : integer
    procedure init(self, price : integer, color : integer, speed : integer)
        self.color := color
        self.price := price
        self.gas := 0
        self.speed := speed
    procedure beep(self)
        write(5555); writeln()
    procedure showStats(self)
        write(self.color); writeln()
        write(self.price); writeln()
        write(self.gas); writeln()
        write(self.speed); writeln()

procedure makeCar(price, color, speed : integer)
    object c : Car
    write(color); writeln()
    c.init(price, color, speed)
    write(c.color); writeln()
    c.showStats()
    c.fillUpTank()
    c.showStats()

program myProg
    var p1, p2 : integer
    var color1, color2 : integer
    var speed : integer
    var tires : integer
    object car1 : Car
    object truck1 : Truck
    makeCar(878, 65, 33)
    p1 := 999; p2 := 888; color1 := 1; color2 := 2
    speed := 90; tires := 10
    car1.init(p1, color1, speed)
    truck1.init(p2, color2, tires)
    writeln();
    car1.showStats();truck1.showStats()
    {car1.fillUpTank(); truck1.fillUpTank()}
    {car1.showStats();truck1.showStats()}
""", dstfn = "AssemblyFiles/RISCV-class3.s", target = 'riscv', debug = False)
# 65
# 65
# 65
# 878
# 0
# 33
# 65
# 878
# 100
# 33

# 1
# 999
# 0
# 90
# 2
# 888
# 0
# 10
```
