### Equivalent Languages

Recall grammar `G₂`. Let `G₂' = (T, N, P, S)` with `T = {a}`, `N = {S}` productions `P`:

    S → ε
    S → Sa

Prove that `L(G₂') = {aⁿ | n ≥ 0}`, which is the same as `L(G₂)`.

*Proof.* This is formally proved by inclusion in both directions. By definition of `L(G₂')`,

    {χ ∈ T* | S ⇒ᐩ χ} ⊆ {aⁿ | n ≥ 0}

means that for every `χ ∈ T*` derivable from `S`, there exists an `n ≥ 0` such that `χ = aⁿ`. This is shown by induction over the length of derivations  from `S`.

- _Base._ A derivation of `χ` of length `1` from `S` can only derive `χ = ε` by the first production. As `ε = a⁰`, the base case holds.
- _Step._ We need to show that each `χ` derivable from `S` in `n + 1` steps, `S ⇒ⁿ⁺¹ χ` is `aⁱ` for some `i ≥ 0`, under the hypothesis that each  `χ` derivable from `S` in `n` steps, `S ⇒ⁿ χ` is `aⁱ` for some `i ≥ 0`. If `χ` is derivable in `n+1` steps, then `S ⇒ Sa ⇒ⁿ χ` and `χ` is `ωa` for some `ω`. Since `ω` is derived from `S` in `n` steps, `ω` is `aⁱ` for some `i ≥ 0`, hence `χ = ωa` is `aⁱ` for some `i ≥ 0`.

The inclusion in the other direction means that every `aⁿ` for `n ≥ 0` can be derived from `S`:

    {aⁿ | n ≥ 0} ⊆ {χ ∈ T* | S ⇒ᐩ χ}

This is shown by induction over `n`.

- _Base._ For `n = 0`, obviously `a⁰ = ε` can be generated by the first production, `S ⇒ᐩ ε`.
- _Step._ Suppose `aⁿ` can be generated, `S ⇒ᐩ aⁿ`. We need to show that `aⁿ⁺¹` can also be generated. This follows from `S ⇒ Sa ⇒ᐩ aⁿa = aⁿ⁺¹`.

Thus we can conclude `L(G₂') = {aⁿ | n ≥ 0}`.
