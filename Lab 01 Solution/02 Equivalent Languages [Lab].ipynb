### Equivalent Languages

Recall grammar `G₂`. Let `G₂' = (T, N, P, S)` with `T = {a}`, `N = {S}` productions `P`:

    S → ε
    S → Sa

Prove that `L(G₂') = {aⁿ | n ≥ 0}`, which is the same as `L(G₂)`.

Your answer here.

*Instructor's Answer:*

By definition of `L(G₂')`, this is equivalent to:

    {χ ∈ T* | S ⇒ᐩ χ} = {aⁿ | n ≥ 0}

This is proved by inclusion in both directions.

    {χ ∈ T* | S ⇒ᐩ χ} ⊆ {aⁿ | n ≥ 0}

means that for every `χ ∈ T*` derivable from `S`, there exists an `n ≥ 0` such that `χ = aⁿ`. This is shown by induction over the length of derivations.

- _Base._ Suppose `χ` is derived directly from `S` by `S ⇒ χ`, which leaves only `χ = ε` according to the first production. Then `χ = a⁰`.
- _Step._ Suppose `χ` is derived from `S` in multiple steps, which leaves only `S ⇒ Sa ⇒* χ` according to the second production. Then `χ` must be `ωa` to be derivable from `Sa`. As the derivation `S ⇒ᐩ ω` is shorter than the derivation of `S ⇒ᐩ χ`, by induction assumption there exists an `n` such that `ω = aⁿ`. Therefore `χ = ωa = aⁿa = aⁿ⁺¹`.

The inclusion in the other direction means that every `aⁿ` for `n ≥ 0` can be derived from `S`:

    {aⁿ | n ≥ 0} ⊆ {χ ∈ T* | S ⇒ᐩ χ}

This is shown by induction over `n`.
- _Base._ Obviously `a⁰ = ε` can be generated by the first production, `S ⇒ᐩ ε`.
- _Step._ Suppose `aⁿ` can be generated, `S ⇒ᐩ aⁿ`. We need to show that `aⁿ⁺¹` can also be generated. This follows from `S ⇒ Sa ⇒ᐩ aⁿa = aⁿ⁺¹`.

Thus we can conclude `L(G₂') = {aⁿ | n ≥ 0}`.
