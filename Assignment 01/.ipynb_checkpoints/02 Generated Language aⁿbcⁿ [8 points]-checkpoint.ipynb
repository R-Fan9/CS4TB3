### Generated Language aⁿbcⁿ

Recall `G₃ = (T, N, P, S)`, where `T = {a, b, c}`, `N = {S}`, and the productions `P` are:

    S → b
    S → aSc

Prove that `L(G₃) = {aⁿbcⁿ | n ≥ 0}`! [8 points]

*Proof.* This is formally proved by inclusion in both directions. By definition of `L(G₃)`,

    {χ ∈ T* | S ⇒ᐩ χ} ⊆ {aⁿbcⁿ | n ≥ 0}

means that for every `χ ∈ T*` derivable from `S`, there exists an `n ≥ 0` such that `χ = aⁿbcⁿ`. This is shown by induction over the length of derivations from `S`.

- _Base._ A derivation of `χ` of length `1` from `S` can only derive `χ = b` by the first production. As `b = a⁰bc⁰`, the base case holds.
- _Step._ We need to show that each `χ` derivable from `S` in `n + 1` steps, `S ⇒ⁿ⁺¹ χ` is `aⁱbcⁱ` for some `i ≥ 0`, under the hypothesis that each  `χ` derivable from `S` in `n` steps, `S ⇒ⁿ χ` is `aⁱbcⁱ` for some `i ≥ 0`. If `χ` is derivable in `n+1` steps, then `S ⇒ aSc ⇒ⁿ χ` and `χ` is `aωc` for some `ω`. Since `ω` is derived from `S` in `n` steps, `ω` is `aⁱbcⁱ` for some `i ≥ 0`, hence `χ = aωc` is `aⁱbcⁱ` for some `i ≥ 0`.

Conversely, every `aⁿ` for `n ≥ 0` can be derived from `S`:

    {aⁿbcⁿ | n ≥ 0} ⊆ {χ ∈ T* | S ⇒ᐩ χ}

This is shown by induction over `n`.

- _Base._ For `n = 0`, obviously `a⁰bc⁰ = b` can be generated by the first production, `S ⇒ᐩ b`.
- _Step._ Suppose `aⁿbcⁿ` can be generated, `S ⇒ᐩ aⁿbcⁿ`. We need to show that `aⁿ⁺¹bcⁿ⁺¹` can also be generated. This follows from `S ⇒ aSc ⇒ᐩ aaⁿbcⁿc = aⁿ⁺¹bcⁿ⁺¹`.

Thus we can conclude `L(G₃) = {aⁿbcⁿ | n ≥ 0}`.
